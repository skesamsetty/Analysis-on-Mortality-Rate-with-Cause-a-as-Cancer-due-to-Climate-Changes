```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
# Import API key
# from config.py import g_key
```

```python
states_df = pd.read_csv("../data/world_country_and_usa_states_latitude_and_longitude_values.csv")
states_df
```

```python
states_coordinates = states_df[["usa_state_code", "usa_state_longitude","usa_state_latitude", "usa_state"]].dropna()
states_coordinates
```

```python
death_rates = pd.read_csv("../data/mortality_rate_by_US_state.csv")
recent_deaths = death_rates.loc[(death_rates["Cause Name"]=="Cancer")&(death_rates["Year"]>1999)]
```

```python
death_df = recent_deaths.groupby(recent_deaths["State"]).mean(["Deaths"])
death_rate_df = death_df[["Deaths"]]
```

```python
temperature_change = pd.read_csv("../data/model_state.csv")
temperature_change
```

```python
highest_temperature_change =temperature_change[["Annual", "STUSAB"]].sort_values(by = "Annual", ascending = False).head()
```

```python
highest_climate = highest_temperature_change.rename(columns = {"STUSAB" : "usa_state_code"})
```

```python
climate_marker_data = highest_climate.merge(states_coordinates , on="usa_state_code", how = "outer").dropna()
```

```python
states_coordinates_death = states_coordinates.rename( columns = {"usa_state" : "State"})
states_coordinates_death
```

```python
death_heatmap_data = death_rate_df.merge(states_coordinates_death, on = "State", how = "outer").dropna()
```

```python
climate_marker_data
```

```python
death =death_heatmap_data.sort_values(by = "Deaths", ascending = False).head()
death
```

```python
locations =death_heatmap_data[["usa_state_latitude", "usa_state_longitude"]]
deaths = death_heatmap_data["Deaths"]
fig = gmaps.figure( center= (46, -5), zoom_level= 2 )
heat_layer = gmaps.heatmap_layer(locations, weights= deaths, 
                                 dissipating=True, max_intensity=100,
                                 point_radius=3, opacity = 1)
fig.add_layer(heat_layer)
fig
```

```python
location = climate_marker_data[["usa_state_latitude", "usa_state_longitude"]]
markers= gmaps.marker_layer(location)

# Display figure
fig.add_layer(markers)
fig
```
